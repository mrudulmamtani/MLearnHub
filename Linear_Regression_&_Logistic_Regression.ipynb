#Linear Regression, Ridge AND Lasso

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
load_boston = '/content/housing.csv'

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
#MEDV is the target variable
boston = pd.read_csv(load_boston, header=None, delimiter=r"\s+", names=column_names)
boston.head()
```

```python
#diving the dataset into independent and dependent
X = boston.iloc[:,:-1]#independent feauture
y = boston.iloc[:,-1]
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg,X,y, scoring = 'neg_mean_squared_error', cv = 5)
mean_mse = np.mean(mse)

print(mean_mse)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

##Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

params = {'alpha':[1e-15,1e-10,1-8,1e-3,1e-2,1,2,3,4,6,5,10,20,30,40,50,60,70,80,]}
ridge_regressor = GridSearchCV(ridge,params,scoring='neg_mean_squared_error', cv = 5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

#Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()

params = {'alpha':[1e-15,1e-10,1-8,1e-3,1e-2,1,2,3,4,6,5,10,20,30,40,50,60,70,80,]}
lasso_regressor = GridSearchCV(lasso,params,scoring='neg_mean_squared_error', cv = 5)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred =ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2 = r2_score(y_pred, y_test)
print(r2)
```

```python
y_pred =lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2 = r2_score(y_pred, y_test)
print(r2)
```

##Importing breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer

df = load_breast_cancer()
X = pd.DataFrame(df['data'], columns = df['feature_names'])
```

```python
X.head()
```

```python
y = pd.DataFrame(df['target'], columns = ['Target'])
y.head()
```

```python
y['Target'].value_counts()
```

```python
#train_test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
params = [{'C': [ 1,5,10]}, {'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=100,max_iter=100)
```

```python
model = GridSearchCV(model1, params, scoring= 'f1', cv = 5)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python

confusion_matrix(y_pred, y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
